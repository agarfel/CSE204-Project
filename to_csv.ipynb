```python
# Debugging
#h5 = hdf5_getters.open_h5_file_read("/mnt/c/Users/Aumit/Desktop/final-proj/MillionSongSubset/data/A/A/A/TRAAADZ128F9348C2E.h5")
#title = hdf5_getters.get_artist_name(h5)
#h5.close()
#print title

# Provide the starting root dir 
# This is different for your particular setup! 
indir = 'MillionSongSubset'
l,t = ([],[])
# Open the CSV file we will write to
with open("output.csv", 'w') as csvfile:
    # Column headers
    # Currently only 4 features being extracted
    # We can add as many as we want, just seperate with commas
    csvfile.write("artist_name,title,release,similar,hottness,artist_terms,artist_terms_weights,familiarity,loudness,tempo")
    #csvfile.write("familiarity, hotttness")
    csvfile.write("\n")

    # Recursively visit each sub-dir till we reach the h5 files
    # Strip punctuation from features that are strings
    for root, dirs, filenames in os.walk(indir):
        for f in filenames:
            if f == '.DS_Store':
                continue
            #if i == 200: break
            #log = open(os.path.join(root, f),'r')
            #print os.path.join(root, f)

            # Use the hd5 wrappers to open the h5 file
            h5 = hdf5_getters.open_h5_file_read(os.path.join(root, f))


            # Get the artist name
            artist_name = hdf5_getters.get_artist_name(h5)
            artist_name = artist_name.decode('utf-8')
            artist_name = artist_name.translate(string.punctuation)  

            # Get the title of the song
            title_song = hdf5_getters.get_title(h5)
            title_bytes = title_song.decode('utf-8')
            title = title_bytes.translate(string.punctuation)  
            
            # Get release
            release_song = hdf5_getters.get_release(h5)
            release_bytes = release_song.decode('utf-8')
            release = release_bytes.translate(string.punctuation)
            
            # Get artist HOTTTNESSSSSS
            hottness = hdf5_getters.get_artist_hotttnesss(h5)

            # Get artist terms
            artist_terms = hdf5_getters.get_artist_terms(h5)
            tmp = artist_terms.tolist()
            artist_terms = ''

            for term in tmp:
                    artist_terms += term.decode('utf-8') +'/'

            artist_term_weights = hdf5_getters.get_artist_terms_weight(h5)
            tmp = artist_term_weights.tolist()
            artist_terms_weights = ''
            for term in tmp:
                    artist_terms_weights += str(term) +'/'
    
            # Get artist familiarity 
            familiarity = hdf5_getters.get_artist_familiarity(h5)
            
            # Get loudness
            loudness = hdf5_getters.get_loudness(h5)
            l.append(loudness)

            # Get tempo
            tempo = hdf5_getters.get_tempo(h5)
            t.append(tempo)
            
            # Get simlar artists
            similar = hdf5_getters.get_similar_artists(h5)
            tmp = similar.tolist()
            similar = ''
            for term in tmp:
                    similar += term.decode('utf-8') +'/'
            
            # Close the h5 file
            h5.close()

            # Write to the CSV file
            csvfile.write(str(artist_name)+ ',' + str(title) +','+str(release)+','+str(similar)+','+str(hottness)+','+str(artist_terms)+','+str(artist_terms_weights)+','+str(familiarity)+','+str(loudness)+','+str(tempo))
            #csvfile.write(str(familiarity) + "," + str(hotttness))
            csvfile.write("\n")

            # Print the current song and arists: 
            #print str(num_songs)
```
