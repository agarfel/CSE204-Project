```python
import pandas as pd
import csv

file_path = "spotify_dataset.csv"

# Open the file and read it line by line
with open(file_path, mode='r', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter=',', quotechar='"', escapechar='\\')
    lines = []
    for row in reader:
        if len(row) == 4:  # Ensure each row has exactly 4 columns
            lines.append(row)

# Convert the cleaned data into a DataFrame
df = pd.DataFrame(lines, columns=["user_id", "artistname", "trackname", "playlistname"])

# Display the first few rows of the dataframe to verify
print(df.head())
```

```python
import pandas as pd
import numpy as np
import build as bd
import random
```

```python
spotifydata = df.copy()
```

```python
data1 = bd.load_data("output.csv", reduce = False, filterz = True)
```

```python
data = data1.copy()
```

```python
#im merging the data sets to only include songs that appear both in our feature dataset and the playlist dataset

data = data.rename(columns={'title': 'trackname', 'artist_name': 'artistname'})
merged_df = pd.merge(spotifydata, data, on=['trackname', 'artistname'])

```

```python
# now im grouping them by playlist so we can read and examine them

merged_filtered = merged_df.groupby('playlistname')
filtered_playlists_dict = {name: group for name, group in merged_filtered if len(group)>=10}
```
